```python
df.isnull().sum()
```

```python
#best practice is to (choose median > mean) use median since its rubust to outliers and use median review rating within each category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#making the format same so that while coding it does not gets in way(using snake casing)
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.info()
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a column age_group
labels = ['Young adult', 'Adult', 'Middle aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#creating another feature = purchase_frequency_days; converting texts in frequency_of_purchases into numbers
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90}
```

```python
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df.to_csv('customer_shopping_behavior.csv', index=False)
```
